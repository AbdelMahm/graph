```python
# supression des liens avec le poids elev
T = TT
#print(T)
GG = G.copy()
d = []
j = 0

while j<len(T):
    (u,v) = T[j][1]
    #print(T[1][1],T[2][1] )
    a = list(GG.neighbors(u))
    b = list(GG.neighbors(v))
    #print((u,v),len(a),len(b))
    if (len(a)>1 and len(b)>1):
        GG.remove_edge(u,v)
        d.append((u,v))
    j = j+1

nx.draw(GG,with_labels=True)
print('')
plt.show()
print('')
d

print(list(nx.connected_components(GG)))
```
